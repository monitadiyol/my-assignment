# 3. Check whether all the zones generate the same amount of sales

H0: All zones generate the same amount of sales

H1: All zones do not generate the same amount of sales

Assumed alpha value=0.05

```python
import scipy.stats as stats
stats.f_oneway(data['Zone - A'],data['Zone - B'],data['Zone - C'],data['Zone - D'])
```

Here the p value is less than assumed alpha value 0.05. Therefore we can reject the null hypothesis

ie, All zones do not generate same amount of sales
