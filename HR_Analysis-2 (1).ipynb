# Data PreProcessing

### NaN Handling

```python
df.isna().sum()
```

### Handling NaN in BasePay

From the previous inferences we found that Base Pay changes with Education

```python
df1=df
df1[df1.Base_pay.isnull()]
```

We can obsrve here that the Base pay is Blank for Education PG 

```python
df2=df1
df3 = df2[df2['Education'] == 'PG']
print(df3)
```

```python
df3
```

```python
#Since the Base Pay is Null in PG Category of Education, we take the skew of PG.

print(df3["Base_pay"].skew())
sns.distplot(df3["Base_pay"])
```

```python
#Replacing the null values with Mean, as the plot is nagatively skewed. 

df1['Base_pay']=df1['Base_pay'].fillna(df1['Base_pay'].mean())
```

### Filling Null Values of Opening Balance

```python
df1=df
print(df1["openingbalance"].skew())
sns.distplot(df1["openingbalance"])
```

```python
#Replacing the null values with Median, as the plot is positively skewed. 

df1['openingbalance']=df1['openingbalance'].fillna(df1['openingbalance'].median())
```

### Filling Null values of Total Sales 

```python
df1['Total_Sales']=df1['Total_Sales'].replace(' ',np.nan,regex = True)
```

```python
df1['Total_Sales']=df1['Total_Sales'].fillna(0)
df1['Total_Sales']
```

```python
print(df1["Total_Sales"].skew())
sns.distplot(df1["Total_Sales"])
```

```python
#Replacing the null values with Median, as the plot is positively skewed. 

df1['Total_Sales']=df1['Total_Sales'].fillna(df1['Total_Sales'].median())
```

```python
s=df.isnull()
sns.heatmap(s)
```

Hence, all  null values are cleared.

# Handling Outlier

```python
mlt.figure(figsize=(15,6))
df1.boxplot()
```

```python
for i in ['Age','Salary','Base_pay','Bonus','Unit_Price','Volume','openingbalance','closingbalance','low']:
    print(i)
    Q1=np.percentile(df1[i],25,interpolation='midpoint')
    Q2=np.percentile(df1[i],50,interpolation='midpoint')
    Q3=np.percentile(df1[i],75,interpolation='midpoint')
    print(Q1, Q2, Q3)

    IQR=Q3-Q1
    print(IQR)   

    low_limit=Q1-1.5*IQR
    upp_limit=Q3+1.5*IQR
    print('LOWER LIMIT = ',low_limit)
    print('UPPER LIMIT = ',upp_limit)

    drop_index = df1[(df1[i] < low_limit) | (df1[i] > upp_limit)].index
    df1.drop(drop_index,inplace=True)
df1 
```
