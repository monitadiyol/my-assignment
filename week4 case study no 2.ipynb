# 1.Read the dataset to the python environment

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("cars_data.csv")
```

# 2.Check for the null values present in the data set

```python
df.isna().sum()
```

# 3.Plot a bar graph of male  vs female buyers participated in the scale

```python
df['Car Gender'].value_counts().plot.bar(figsize=(12,4))
```

# 4.Find the top 5 cars based on their scale price

```python
Make=df.groupby('Make')
Make.first()
```

```python
plt.figure
Make['Sale Price'].sum().nlargest(5).sort_values(ascending=False).plot.bar(figsize=(12,4),title='top five cars')
```

# 5.Find the least 5cars based on this resell price

```python
plt.figure
Make['Resell Price'].sum().nsmallest(5).sort_values(ascending=True).plot.bar(figsize=(12,4),title='least five cars')
```
